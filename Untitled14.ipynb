```python
import kagglehub
import pandas as pd

# Download latest version
path = kagglehub.dataset_download("kreeshrajani/3k-conversations-dataset-for-chatbot")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(path+"/Conversation.csv")
df
```

```python
question = df['question'].to_list()
answers = df['answer'].to_list()
datasets = []
for i in range(len(question)) :
  data = question[i] + " " + answers[i]
  datasets.append(data)
```

```python
x_train = ["sos " + seq for seq in datasets]
y_train = [seq + " eos" for seq in datasets]
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(x_train)
tokenizer.fit_on_texts(y_train)
len(tokenizer.word_index)
```

```python
import joblib
joblib.dump(tokenizer,"tokenizer.pkl")
```

```python
@keras.utils.register_keras_serializable()
class LCM (keras.layers.Layer) :
  def __init__(self,d_model,drop_out : float= 0.1,**kwargs)  :
    super(LCM,self).__init__(**kwargs)
    self.d_model = d_model
    self.drop_rate = drop_out
    self.dropout = keras.layers.Dropout(self.drop_rate)
    self.step1 = keras.layers.Dense(d_model)
    self.step2 = keras.layers.Dense(d_model)
    self.mag_nitude = keras.layers.Dense(d_model,activation=keras.activations.tanh)
    self.norm = keras.layers.LayerNormalization()

  def call(self,x,training=False) :
    normlogits = self.norm(x)
    step1 = self.step1(normlogits)
    step1 = keras.activations.gelu(step1,approximate=True)
    step2 = self.step2(normlogits)
    step2 = keras.activations.gelu(step2,approximate=True)
    laten = step1 + step2
    laten = self.dropout(laten,training=training)
    laten = self.mag_nitude(laten)
    return x + laten


  def get_config(self) :
    config = super(LCM,self).get_config()
    config.update({
        'd_model' : self.d_model,
        'drop_rate' :self.drop_rate
    })
    return config

  @classmethod
  def from_config(cls,config) :
    return cls(**config)
```

```python
@keras.utils.register_keras_serializable()
class LCTBlock (keras.layers.Layer) :
  def __init__ (self,d_model,drop_rate,**kwargs) :
    super(LCTBlock,self).__init__(**kwargs)
    self.d_model = d_model
    self.droprate= drop_rate
    self.LCM = LCM(self.d_model,self.droprate)
    self.MHA = keras.layers.MultiHeadAttention(num_heads=4,key_dim=(self.d_model//4),dropout=self.droprate)
    self.layernorm = keras.layers.LayerNormalization()

  def call(self,x,training=False) :
    norm = self.layernorm(x)
    attn = self.MHA(norm,norm,norm,use_causal_mask=True,training = training)
    x+=attn
    return self.LCM(x)

  def get_config(self) :
    config = super(LCTBlock,self).get_config()
    config.update({
        'd_model' : self.d_model,
        'drop_rate' : self.droprate
    })
    return config

  @classmethod
  def from_config(cls,config) :
    return cls(**config)
```

```python
@keras.utils.register_keras_serializable()
class LatenConnectionTransformers (keras.Model) :
  def __init__ (self,vocab_size,d_model,drop_rate = 0.1,maxpos=100,**kwargs) :
    super(LatenConnectionTransformers,self).__init__(**kwargs)
    self.vocab_size = vocab_size
    self.d_model = d_model
    self.drop_rate = drop_rate
    self.maxpos = maxpos
    self.ffn = keras.Sequential([
        keras.layers.Dense(d_model*4,activation='relu'),
        keras.layers.Dense(d_model)
    ],name="FFN-BottleNeck")
    self.embedding = keras.layers.Embedding(self.vocab_size,self.d_model)
    self.pos_learn = keras.layers.Embedding(self.maxpos,self.d_model)
    self.block1 = LCTBlock(self.d_model,self.drop_rate)
    self.block2 = LCTBlock(self.d_model,self.drop_rate)
    self.block3 = LCTBlock(self.d_model,self.drop_rate)
    self.Linear = keras.layers.Dense(self.vocab_size)

  def call(self,x,training=True) :
    x = self.embedding(x)
    x *= tf.sqrt(tf.cast(self.d_model,tf.float32))
    pos = tf.range(tf.shape(x)[1])
    pos = self.pos_learn(pos)
    pos = tf.expand_dims(pos,axis=0)
    x+=pos
    x = self.block1(x,training=training)
    x = self.block2(x,training=training)
    x = self.block3(x,training=training)
    x = self.ffn(x)
    x = self.Linear(x)
    return x

  def get_config(self) :
    config = super(LatenConnectionTransformers,self).get_config()
    config.update({
        'd_model' : self.d_model,
        'drop_rate' : self.drop_rate,
        'maxpos' : self.maxpos,
        'vocab_size' : self.vocab_size
    })
    return config

  @classmethod
  def from_config(cls,config) :
    return cls(**config)
```

```python
x_train = tokenizer.texts_to_sequences(x_train)
y_train = tokenizer.texts_to_sequences(y_train)
x_train = pad_sequences(x_train,padding='post',maxlen=174)
y_train = pad_sequences(y_train,padding='post',maxlen=174)
```

```python
chatbotModel = LatenConnectionTransformers(vocab_size=len(tokenizer.word_index)+1,d_model=64,drop_rate=0.1,maxpos=200)
chatbotModel.compile(optimizer=keras.optimizers.AdamW(learning_rate=5e-4,clipnorm=1.0),
                     loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                     metrics=['accuracy'])
```

```python
eval = chatbotModel.fit(x_train,y_train,epochs=100,batch_size=64,verbose=1)
```

```python
import matplotlib.pyplot as plt
plt.title("LCT Chatbot AI training bechmark ")
plt.xlabel("epoch")
plt.ylabel("values")
plt.plot(eval.history['accuracy'],label='accuracy',color='green')
plt.plot(eval.history['loss'],label='loss',color='red')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
```

```python
def generate_text(prompt, max_len=80, temperature=1.0):

    seq = [tokenizer.word_index["sos"]] + tokenizer.texts_to_sequences([prompt])[0]
    word = " "

    for _ in range(max_len):
        pad = pad_sequences([seq], maxlen=174, padding="post")

        logits = chatbotModel(pad, training=False)
        logits = logits[:, len(seq)-1, :]


        logits = logits / temperature
        probs = tf.nn.softmax(logits).numpy()[0]

        next_id = np.random.choice(len(probs), p=probs)

        seq.append(next_id)

        if next_id == tokenizer.word_index["eos"]:
            break
        word += tokenizer.index_word[next_id] + " "

    return word
```
